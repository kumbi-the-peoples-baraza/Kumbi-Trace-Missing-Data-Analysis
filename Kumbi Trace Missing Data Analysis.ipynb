```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
file_path = '.ipynb_checkpoints/Anonymous Data.xlsx'
data = pd.read_excel(file_path)
```

```python
# Display the first few rows of the dataframe
data.head()
```

```python
#display all the rows of a DataFrame
# Set display options to show all rows
pd.set_option('display.max_rows', None)
# Display the DataFrame
print(data)
```

```python
#DATA ANALYSIS BASED ON;
#Gender Distribution: Pie chart showing the distribution of missing persons by gender.
#Age Distribution: Histogram showing the distribution of ages of the missing persons.
#Status of Missing Persons: Bar chart showing the number of people found alive, found dead, and still missing.
#Last Seen Location Distribution:Bar chart showing the distribution of the last seen locations of missing persons.
#Missing Dates: Line or bar chart showing the number of people reported missing over time.
#Image Availability: Pie chart showing the percentage of cases with and without images.
#Summary Statistics:
#Gender Distribution:
#Provides a count of each gender.
#Age Distribution:
#Includes count, mean, standard deviation, min, 25%, 50%, 75%, and max.
#Status of Missing Persons:
#Provides a count of each status.
#Image Availability:
#Provides a count of "Yes" and "No".
#Missing Dates:
#Includes count, mean, standard deviation, min, 25%, 50%, 75%, and max for the resampled monthly counts.
#Last Seen Location Distribution:
#Provides a count of each location.
```

```python
# Preparing data for visualizations
```

```python
# Gender Distribution
gender_distribution = data['GENDER'].value_counts()
```

```python
# Age Distribution
age_distribution = data['AGE']
```

```python
# Status of Missing Persons
status_distribution = data[['FOUND ALIVE', 'FOUND DEAD']].apply(lambda x: 'Found Alive' if x['FOUND ALIVE'] == 'YES' else ('Found Dead' if x['FOUND DEAD'] == 'YES' else 'Still Missing'), axis=1).value_counts()
```

```python
# Last Seen Location Distribution
location_distribution = data['LAST SEEN LOCATION'].value_counts()
```

```python
# Image Availability
image_availability = data['Image available'].value_counts()
```

```python
# Missing Dates
missing_dates = pd.to_datetime(data['MISSING DATE'], format='%d/%m/%Y').value_counts().sort_index()
```

```python
# Plotting each graph separately
```

```python
# Gender Distribution
plt.figure(figsize=(8, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.savefig('gender_distribution.svg')
plt.show()
```

```python
#Summary Statistics of Gender Distribution
summary_gender = gender_distribution.describe()
```

```python
print("Summary Statistics for Gender Distribution:")
print(summary_gender)
```

```python
 # Age Distribution with a maximum age of 55 years and a scale of 5 years
plt.figure(figsize=(8, 6))
plt.hist(age_distribution, bins=range(0, 56, 5), edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.xticks(range(0, 56, 5))
plt.savefig('age_distribution.svg')
plt.show()
```

```python
#Summary statistics of Age Distribution
summary_age = age_distribution.describe()
```

```python
print("\nSummary Statistics for Age Distribution:")
print(summary_age)
```

```python
# Status of Missing Persons
plt.figure(figsize=(8, 6))
plt.bar(status_distribution.index, status_distribution.values)
plt.title('Status of Missing Persons')
plt.xlabel('Status')
plt.ylabel('Number of People')
plt.savefig('status_of_missing_persons.svg')
plt.show()
```

```python
#Summary statistics of Status of Missing Persons
summary_status = status_distribution.describe()
```

```python
print("\nSummary Statistics for Status of Missing Persons:")
print(summary_status)
```

```python
plt.figure(figsize=(12, 8))
location_distribution.sort_values().plot(kind='barh')
plt.title('Last Seen Location Distribution')
plt.xlabel('Number of People')
plt.ylabel('Location')
plt.savefig('last_seen_location_distribution.svg')
plt.show()
```

```python
#Summary statistics of Last Seen Location Distribution
summary_location = location_distribution.describe()
```

```python
print("\nSummary Statistics for Last Seen Location Distribution:")
print(summary_location)
```

```python
plt.figure(figsize=(12, 6))
missing_dates.plot(kind='line', marker='o', color='blue')
plt.title('Missing Dates')
plt.xlabel('Date')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('missing_dates.svg')
plt.show()
```

```python
#Summary statistics of Missing Dates
summary_missing_dates = missing_dates.describe()
```

```python
print("\nSummary Statistics for Missing Dates:")
print(summary_missing_dates)
```

```python
# Image Availability
plt.figure(figsize=(8, 6))
plt.pie(image_availability, labels=image_availability.index, autopct='%1.1f%%', startangle=140)
plt.title('Image Availability')
plt.savefig('image_availability.svg')
plt.show()
```

```python
#Summary Statistics of Image Availability
summary_image_availability = image_availability.describe()
```

```python
print("\nSummary Statistics for Image Availability:")
print(summary_image_availability)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and 'LAST SEEN DATE' is the column with the dates
missing_dates = pd.to_datetime(data['LAST SEEN DATE'], errors='coerce')

# Resample the data by month and count the occurrences
missing_dates = missing_dates.dt.to_period('M').value_counts().sort_index()

# Convert PeriodIndex to datetime for plotting
missing_dates.index = missing_dates.index.to_timestamp()

# Plotting
plt.figure(figsize=(12, 6))
missing_dates.plot(kind='line', marker='o', color='blue')
plt.title('Missing Dates')
plt.xlabel('Month')
plt.ylabel('Number of People')
plt.grid(True)

# Customize x-axis to show specific months
xticks = pd.date_range(start=missing_dates.index.min(), end=missing_dates.index.max(), freq='3MS')  # Every 3 months
plt.xticks(ticks=xticks, labels=[d.strftime('%b %Y') for d in xticks], rotation=45)

plt.tight_layout()  # Adjust layout to prevent label cut-off
plt.show()
```

```python
#LAST SEEN DATES DISTRIBUTION
# Make sure the 'LAST SEEN DATE' is in datetime format
data['LAST SEEN DATE'] = pd.to_datetime(data['LAST SEEN DATE'])
```

```python
# Filter the data for the year 2024
data_2024 = data[data['LAST SEEN DATE'].dt.year == 2024]
```

```python
# Group the data by month to count occurrences
monthly_counts = data_2024['LAST SEEN DATE'].dt.month.value_counts().sort_index()
```

```python
# Ensure that all months are represented, even if some have zero counts
full_months = np.arange(1, 13)
monthly_counts_full = pd.Series(0, index=full_months)
monthly_counts_full.update(monthly_counts)
```

```python
# Define month labels
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
```

```python
# Plotting the distribution with corrected counts
plt.figure(figsize=(10, 6))
plt.bar(months, monthly_counts_full, color='blue')
plt.title('Distribution of Last Seen Dates in 2024')
plt.xlabel('Month')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
# Save the plot as a PDF
plt.savefig('last_seen_distribution_2024.pdf')
# Show plot
plt.show()
```
